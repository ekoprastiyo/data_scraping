```python
import numpy as np
import pandas as pd 
from googletrans import Translator 
```

```python
path = 'perpusnas_google_review-2023_01_27_17_57_20.csv'
df = pd.read_csv(path)
```

```python
df
```

```python
reviews = df[df['review'].notnull()][:-2].reset_index()
```

```python
reviews
```

```python
#create our translator object 
translator = Translator()

reviews['en_review'] = reviews['review'].apply(lambda x: translator.translate(x, src='auto', dest='en').text)
```
