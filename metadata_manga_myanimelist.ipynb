```python
def cleaning_data(meta):
    all_part = []
    for met in meta:
        part = []
        for x in met:
            if ':' in x:
                dicts, vals = x.split(': ', 1)
                part.insert(len(part), dicts)
                part.insert(len(part), vals)
            else:
                continue
        all_part.append(part)
        
    lis_dct = []
    for x in all_part:
        res_dct = {x[i]: x[i + 1] for i in range(0, len(x), 2)}
        lis_dct.append(res_dct)
    df = pd.DataFrame.from_records(lis_dct)
    return df
```

```python
ootani = cleaning_data(meta)
```
